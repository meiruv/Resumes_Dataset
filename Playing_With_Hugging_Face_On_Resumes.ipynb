# Classifying with zero-shot

```python
!pip install datasets transformers[sentencepiece]
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
```

```python
candidate_labelss = resume_df['Category'].unique()
candidate_labelss
```

```python

classifier(
    resume_df['Resume'],
    candidate_labels=["education", "politics", "business"],
)
```

```python
classifier(
    resume_df.loc[0,'Resume'],
    candidate_labels=candidate_labelss,
)
```

```python
x = pd.DataFrame(classifier(
    resume_df.loc[0,'Resume'],
    candidate_labels=candidate_labelss,
))
```
